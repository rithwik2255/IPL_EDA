#Most Wins

```python
temp=pd.DataFrame({'Most Wins':df['Match_winner']})
count_wins = temp.value_counts()
print(count_wins)
labels = [X[0] for X in count_wins.keys()]
ax = plt.pie(x=count_wins,autopct="%.1f%%",labels=labels)
plt.title('Most wins', fontsize=17)
plt.show()
```

#Toss Decision

```python
teams = df1['Toss_winner'].unique()
decision_making = pd.DataFrame([],columns=['Toss_winner','Toss_decision','Times'])
for id, element in enumerate(teams):
    temp_bat = df1[(df1['Toss_winner']==element) & (df1['Toss_decision']=='Bat')]
    temp_field = df1[(df1['Toss_winner']==element) & (df1['Toss_decision']=='Field')]

    #append to decision_making
    decision_making = decision_making.append({'Toss_winner':element, 'Toss_decision':'Bat', 'Times':temp_bat['Toss_winner'].count()},ignore_index=True)
    decision_making = decision_making.append({'Toss_winner':element, 'Toss_decision':'Field', 'Times':temp_field['Toss_winner'].count()},ignore_index=True)
```

```python
decision_making
```

```python
sns.catplot(x="Toss_winner", y="Times", hue="Toss_decision", data=decision_making, height=5, aspect=2)
plt.xticks(rotation=90)
plt.title("Toss decision of teams")
plt.xlabel("IPL Teams")
plt.ylabel("Toss Decision")
plt.show()
```
